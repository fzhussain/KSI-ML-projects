### Applying K means Clustering

```python
# Import KMeans from SciKit Learn
from sklearn.cluster import KMeans
```

```python
# Create an instance of a K Means model with 2 clusters.
kmeans = KMeans(n_clusters=2)
```

```python
# Fit the model to all the data except for the Private label.
kmeans.fit(data.drop('Private', axis=1))
```

```python
means=kmeans.cluster_centers_
print(means)
```

### Evaluation

```python
data['Private'] = data.Private.astype("category").cat.codes
# astype("category"): add a new categorical column to an existing table 
# .cat.codes: Return Series of codes as well as the index.
```

```python
data.Private
```

```python
data.head()
```

#### There is no perfect way to evaluate clustering if we donâ€™t have the labels, however, we do have the labels, so we take advantage of this to evaluate our clusters.
#### Create a new column for df called â€˜Clusterâ€™, which is a 1 for a Private school, and a 0 for a public school.

```python
def converter(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
```

```python
data['Cluster'] = data['Private'].apply(converter)
```

#### Creating a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(data.Private, kmeans.labels_))
print(classification_report(data.Private, kmeans.labels_))
```

```python
print(f'Accuracy:{accuracy_score(data.Private, kmeans.labels_)}')
```
